```python
import pandas as pd

df = pd.read_csv('D:\Conformalized_Quantile_Regression\LUCAS_2015_features_V2.csv')
```

```python
df
```

```python
# Define the number of columns in each group
group_size = 61

# Calculate the number of groups
num_groups = 14

# number of the spectral bands
num_bands = 11 

# Initialize lists to store the results
max_values = []
min_values = []
average_values = []

# Loop through the groups and calculate max, min, and average for each
for i in range(num_groups):
    start_idx = 4 * num_bands + i * group_size + i  #4 is the number of neighborhood pixels that are saved
    end_idx = 4 * num_bands + (i+1) * group_size
    
    group = df.iloc[:, start_idx:end_idx]
    
    # Calculate max, min, and average for each group of columns
    max_values.append(group.max(axis=1))
    min_values.append(group.min(axis=1))
    average_values.append(group.mean(axis=1))

# Calculate the differences between max and min values for each group
diff_values = [max_val - min_val for max_val, min_val in zip(max_values, min_values)]

# Create new DataFrames for max, min, and average values
diff_df = pd.concat(diff_values, axis=1)
average_df = pd.concat(average_values, axis=1)

# Rename the columns of the new DataFrames
diff_df.columns = [f"diff_{i}" for i in range(1, num_groups + 1)]
average_df.columns = [f"average_{i}" for i in range(1, num_groups + 1)]

# Concatenate the results into a single DataFrame
result_df = pd.concat([diff_df, average_df], axis=1)

# Save the result to a new CSV file
# result_df.to_csv('result.csv', index=False)
```
